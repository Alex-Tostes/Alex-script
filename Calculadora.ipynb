```python
# Requisito 1: Utilizar o comando 'input' para receber ao menos 2 números de entrada
print("Calculadora Simples")

# Solicita e recebe o primeiro número do usuário
# Requisito 2: Converter os valores recebidos para número flutuante (float)
try:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
except ValueError:
    print("Entrada inválida. Por favor, digite números válidos.")
    # Encerra o programa se a entrada inicial for inválida
    exit()

print("\nSelecione a operação:")
print("1 - Adição (+)")
print("2 - Subtração (-)")
print("3 - Multiplicação (*)")
print("4 - Divisão (/)")

# Requisito 4: Adicionar um laço de repetição ou uma condicional.
# Utilizaremos uma condicional (if/elif/else) para permitir ao usuário escolher a operação.
escolha = input("Digite a opção (1/2/3/4): ")

# Inicializa a variável resultado
resultado = None
operacao_texto = ""

# Condicional para realizar a operação escolhida
if escolha == '1':
    # Requisito 3: Implementar a operação de Adição
    resultado = num1 + num2
    operacao_texto = "Adição"
elif escolha == '2':
    # Requisito 3: Implementar a operação de Subtração
    resultado = num1 - num2
    operacao_texto = "Subtração"
elif escolha == '3':
    # Requisito 3: Implementar a operação de Multiplicação
    resultado = num1 * num2
    operacao_texto = "Multiplicação"
elif escolha == '4':
    # Requisito 3: Implementar a operação de Divisão
    if num2 != 0:
        resultado = num1 / num2
        operacao_texto = "Divisão"
    else:
        # Condição especial para evitar divisão por zero
        print("Erro: Divisão por zero não é permitida.")
        exit() # Encerra o programa após o erro
else:
    print("Opção inválida. Por favor, escolha uma das opções (1, 2, 3 ou 4).")

# Requisito 5: Utilizar o comando 'print' para exibir o resultado
if resultado is not None:
    print(f"\n--- Resultado da {operacao_texto} ---")
    print(f"O resultado de {num1} e {num2} é: {resultado}")
    print("---------------------------------")
```
